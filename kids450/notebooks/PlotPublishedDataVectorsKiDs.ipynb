## Checking Covariance Values

```python
def get_published_cov(cov_matrix,dt,i,j):
    
    if dt=='xip':
        theta_bins = 7 
        index1i = i*13
        index2i = i*13+7
        index1j = j*13
        index2j = j*13+7
    if dt=='xim':
        theta_bins = 6
        index1i = i*13+7
        index2i = i*13+7+6
        index1j = j*13+7
        index2j = j*13+7+6
    
    
    cov = cov_matrix[index1j:index2j,index1i:index2i]
    return cov
```
