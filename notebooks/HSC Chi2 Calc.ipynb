# HSC $\chi^{2}$ Fit

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_theme()
```

```python
twopoint_data = np.loadtxt('hsc-y1-test/data_vector/2pt_data.txt')
cov = np.loadtxt('hsc-y1-test/data_vector/2pt_covariance.txt')
twopoint_theory = np.loadtxt('hsc-y1-test/data_vector/2pt_theory.txt')
```

```python
def Hartlap(nd, ns):
    return (ns-nd-2.)/(ns-1)
```

```python
def chi2(cov,dv,theory):
    
    residuals = dv-theory
    invcov = (Hartlap(len(dv), 2268))*np.linalg.inv(cov)
    
    chi = (np.dot(np.dot(dv-theory, invcov), dv-theory))
    
    
    return chi
```

```python
chi2(cov,twopoint_data,twopoint_theory)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(15,5))
plt.title('Best Fit Full Data Vector',fontsize=20)
plt.errorbar(np.arange(len(twopoint_data)),twopoint_data,yerr=np.sqrt(np.diag(cov)), c='k', lw=1, marker='.', fmt='.', capthick=0.8, capsize=2.2)
plt.plot(np.arange(len(twopoint_data)), twopoint_theory, label='Best fit')
plt.legend()
```
