# Plotting the Survey's N(z)'s

```python
import matplotlib.pyplot as plt
import numpy as np
import sacc
import seaborn as sns
```

## load the n(z)'s for each survey

```python
# DES
data_dir = '/global/cscratch1/sd/elp25/TXPipe_data/desy1/outputs/'
des_data = sacc.Sacc.load_fits(data_dir+'summary_statistics_publishedcov_linearmid.sacc')
```

```python
# KiDS1000
data_dir = '/global/cscratch1/sd/elp25/TXPipe_data/LATEST/kids1000/'
kids_data = sacc.Sacc.load_fits(data_dir+'summary_statistics_publishedcov_linearmid.sacc')
```

```python
# HSC
data_dir = '/global/cscratch1/sd/elp25/TXPipe_data/LATEST/HSC/'
hsc_data = sacc.Sacc.load_fits(data_dir+'summary_statistics_publishedcov_linearmid.sacc')
```

## plot the n(z)'s on the same plot

```python
sns.color_palette("Blues")
```

```python
import matplotlib.pyplot as plt
import numpy as np
import warnings
warnings.filterwarnings('ignore')

# DES 
for i in range(4):
    ax1 = plt.subplot(311)
    ax1.set(ylabel='n(z)')
    ax1.tick_params(direction="in")
    tr = des_data.get_tracer('source_%d' %i)
    plt.plot(tr.z,tr.nz,label='DES',color=sns.color_palette("Blues")[i])
    plt.setp(ax1.get_xticklabels(), fontsize=6)
    if i==3:
        plt.text(x=2,y=2.2,s='DESY1')

# KiDS1000
for i in range(5):
    ax2 = plt.subplot(312, sharex=ax1)
    ax2.set(ylabel='n(z)')
    ax2.tick_params(direction="in")
    j = i+1
    tr = kids_data.get_tracer('source_%d' %j)
    plt.plot(tr.z,tr.nz,label='KiDS1000',color=sns.color_palette("Greens")[i])
    plt.setp(ax2.get_xticklabels(), fontsize=6)
    if j==5:
        plt.text(x=2.0,y=.14,s='KiDS1000')

# HSC
for i in range(4):
    ax3 = plt.subplot(313, sharex=ax1)
    ax3.set(ylabel='n(z)')
    ax3.tick_params(direction="in")
    j = i+1
    tr = hsc_data.get_tracer('source_%d' %j)
    plt.plot(tr.z,tr.nz,label='HSC',color=sns.color_palette("Reds")[i])
    plt.setp(ax3.get_xticklabels(), fontsize=10)
    plt.xlabel('z')
    plt.xlim(0,3)
    if i==3:
        plt.text(x=2,y=2.5,s='HSC')
plt.show()
```
