# Survey FootPrints

```python
import h5py as h
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
import matplotlib
import numpy as np
```

```python
des = h.File('/global/cscratch1/sd/elp25/txpipe-reanalysis/data/desy1/data/shear_catalog_desy1.h5','r+')
```

```python
hsc = h.File('/global/cscratch1/sd/elp25/txpipe-reanalysis/data/hsc/data/hsc_shear_catalog.hdf5','r+')
```

```python
kids = h.File('/global/cscratch1/sd/elp25/txpipe-reanalysis/data/shear_catalog_kids1000.h5','r+')
```

```python
cut_size = 1000

des_ra = des['shear/ra'][()][0:cut_size]
des_dec = des['shear/dec'][()][0:cut_size]

hsc_ra = hsc['shear/ra'][()][0:cut_size]
hsc_dec = hsc['shear/dec'][()][0:cut_size]

kids_ra = kids['shear/ra'][()][0:cut_size]
kids_dec = kids['shear/dec'][()][0:cut_size]
```

```python
colors = [sns.color_palette("Blues")[2],sns.color_palette("Greens")[2],sns.color_palette("Reds")[2]]
```

```python
plt.style.use('seaborn-whitegrid')


import matplotlib
dpi = 125

matplotlib.rcParams['figure.dpi']= dpi
plt.rc('font', family='serif', size=19)
matplotlib.rcParams['text.usetex']= True 






matplotlib.rc('xtick', labelsize=19) 
matplotlib.rc('ytick', labelsize=19) 

from matplotlib.lines import Line2D
custom_lines = [Line2D([0], [0], lw=4, color=colors[0], label='DES-Y1'),
                Line2D([0], [0], lw=4, color=colors[1], label='HSC-Y1'),
                Line2D([0], [0], lw=4, color=colors[2], label='KiDS-1000')]


plt.figure(figsize=(12,5))
plt.scatter(des_ra,des_dec,alpha=0.02,s=0.02,label='DES-Y1',color=colors[0])
plt.scatter(hsc_ra,hsc_dec,alpha=0.02,s=0.02,label='HSC',color=colors[1])
plt.scatter(kids_ra,kids_dec,alpha=0.02,s=0.02,label='KiDS-1000',color=colors[2])
plt.xlim(360,0)
plt.legend(loc='best',handles=custom_lines,frameon=True,fontsize=19)
plt.xlabel('RA',fontsize=19)
plt.ylabel('Dec',fontsize=19)
plt.show()
```
